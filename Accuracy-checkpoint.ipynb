```python
import numpy as np
import pandas as pd
import sklearn
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score
from sklearn.ensemble import IsolationForest
```

```python
#Training data
df_train=pd.read_csv(r"C:\Users\deeks\Desktop\SET\Train.csv")
```

```python
#Testing data
df_test=pd.read_csv(r"C:\Users\deeks\Desktop\SET\Test.csv")
```

```python
least_crowded=df_train[df_train.traffic_volume<=5000]
heavily_crowded=df_train[df_train.traffic_volume>5000]
```

```python
print(least_crowded.shape)
print(heavily_crowded.shape)
```

```python
f,(ax1,ax2)=plt.subplots(2,1,sharex=True)
f.suptitle("Traffic Volume")
bins=100
ax1.hist(least_crowded.traffic_volume,bins=bins)
ax1.set_title('Least Crowded')
ax2.hist(heavily_crowded.traffic_volume,bins=bins)
ax2.set_title('Heavily Crowded')
plt.xlabel('Traffic')
plt.ylabel('Volume')
plt.xlim((100,10000))
plt.yscale('log')
plt.show();
```

```python
df_train.groupby('traffic_volume').mean()
```

```python
least_crowded_sample=least_crowded.sample(n=7926)
least_crowded_sample
```

```python
new_dataset = pd.concat([least_crowded_sample,heavily_crowded],axis=0)
new_dataset
```

```python
count_newclass=pd.value_counts(new_dataset['traffic_volume'],sort=True)
count_newclass
count_newclass.plot(kind="bar")
labels=("least crowded","heavily crowded")
plt.title("Traffic Volume Distribution")
plt.xticks(range(2),labels)
plt.xlabel("class")
plt.ylabel("Frequency")
```

```python
new_dataset.groupby('traffic_volume').mean()
```

```python
x=new_dataset.drop(columns=["traffic_volume","date_time","is_holiday","weather_type","weather_description"],axis=1)
y=new_dataset["traffic_volume"]
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
#training the logistic regression model with training data
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Model Evaluation
#Accurate train 
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print("Acurrancy of training data:",training_data_accuracy)
```

```python
#Accurate in test
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
print("Acurrancy of test data:",test_data_accuracy)
```
